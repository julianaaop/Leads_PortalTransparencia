```python
from selenium import webdriver
import pyautogui
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from pandas import DataFrame
from selenium.webdriver.common.by import By

def left(s, amount):
    return s[:amount]    

driver = webdriver.Chrome(executable_path=r"C:\chromedriver\chromedriver.exe")
driver.get('https://portal.seuma.fortaleza.ce.gov.br/fortalezaonline/portal/portaltransparencia.jsf')
time.sleep(4)

#if(Bairro!=''):
    #search=driver.find_element(By.NAME,"tvTransparencia:formPortalTransparenciaEndereco:bairro")
    #search.send_keys(Bairro)
#else:
search=driver.find_element(By.NAME,"tvTransparencia:formPortalTransparenciaEndereco:logradouro")
search.send_keys(LOGRADOURO)
    
search.send_keys(Keys.RETURN)
time.sleep(30)    
  
A00=driver.find_element(By.ID,"tvTransparencia:tab01")
A0=A00.find_element(By.ID,"tvTransparencia:formPortalTransparenciaEndereco")
A1=A0.find_element(By.ID,"tvTransparencia:formPortalTransparenciaEndereco:tvResultados")
A2=A1.find_element(By.CLASS_NAME,"ui-tabs-nav.ui-helper-reset.ui-helper-clearfix.ui-widget-header.ui-corner-all")
A3=A2.find_element(By.CLASS_NAME,'ui-state-default.ui-corner-top')
Button=A3.find_element(By.XPATH,'//*[@id="tvTransparencia:formPortalTransparenciaEndereco:tvResultados"]/ul/li[2]/a')
Button.click()

time.sleep(30)
RUA=1 #RUA DA LISTA
A4=A1.find_element(By.CLASS_NAME,"ui-tabs-panels")
A5=A4.find_element(By.CLASS_NAME,"ui-tabs-panel.ui-widget-content.ui-corner-bottom.ui-helper-hidden")
A6=A5.find_element(By.CLASS_NAME,"panel.panel-primary")
A7=A6.find_element(By.ID,"tvTransparencia:formPortalTransparenciaEndereco:tvResultados:dtListaLogradouros")
A8=A7.find_element(By.CLASS_NAME,"ui-datatable-tablewrapper")
A9=A8.find_element(By.ID,"tvTransparencia:formPortalTransparenciaEndereco:tvResultados:dtListaLogradouros_data")

if(RUA%2==0):
    A10=A9.find_elements(By.CLASS_NAME,"ui-widget-content.ui-datatable-odd.ui-datatable-selectable")[int((RUA/2)-1)]

else:
    A10=A9.find_elements(By.CLASS_NAME,"ui-widget-content.ui-datatable-even.ui-datatable-selectable")[int(((RUA+1)/2)-1)]
    
Button2=A10.find_element(By.ID,"tvTransparencia:formPortalTransparenciaEndereco:tvResultados:dtListaLogradouros:"+str(RUA-1)+":row2")
Button2.click()

time.sleep(30)    

B00=driver.find_element(By.ID,"page-content-wrapper")
B0=B00.find_element(By.CLASS_NAME,"col-sm-10.main")
B1=B0.find_element(By.CLASS_NAME,"content-wrapper")
B2=B1.find_element(By.CLASS_NAME,"content")
B3=B2.find_element(By.ID,"fragIncludes")

time.sleep(5) 

B4=B3.find_element(By.ID,"j_idt124")
B5=B4.find_element(By.ID,"formDetalhePortalTransparencia")

time.sleep(5) 

B6=B5.find_element(By.CLASS_NAME,"modal.fade.in")
B7=B6.find_element(By.CLASS_NAME,"modal-dialog")
B8=B7.find_element(By.CLASS_NAME,"modal-content")
B9=B8.find_element(By.CLASS_NAME,"modal-body")
B10=B9.find_element(By.ID,"formDetalhePortalTransparencia:frgTipoDocumentos")
B11=B10.find_element(By.ID,"formDetalhePortalTransparencia:j_idt126")
B12=B11.find_element(By.ID,"formDetalhePortalTransparencia:j_idt127")


B13=B12.find_elements(By.CLASS_NAME,"ui-widget-content")[1]
B14=B13.find_element(By.CLASS_NAME,"ui-panelgrid-cell")
B15=B14.find_element(By.CLASS_NAME,"ui-selectonelistbox.ui-inputfield.ui-widget.ui-widget-content.ui-corner-all.caixaAlta")
B16=B15.find_element(By.CLASS_NAME,"ui-selectlistbox-listcontainer")
B17=B16.find_element(By.CLASS_NAME,"ui-selectlistbox-list")
items = B17.find_elements(By.TAG_NAME,"li")

i=0
for item in items:
    texto=left(item.text,24)
    if(SERVICO in texto): 
        print('ENCONTRADO')
        a=i
    else:
        i=i+1

    
Button3=B17.find_elements(By.TAG_NAME,"li")[a]
Button3.click()

time.sleep(30) 

LEADS= []

if(SERVICO=='PLANO DE RESIDUOS'):
    A='Planos'
elif(SERVICO=='ALVARÁ DE FUNCIONAMENTO'):
    A='AlvarasFuncionamento'
elif(SERVICO=='LICENÇA SANITÁRIA'):
    A='LicencasSanitarias'
    
for i in range(1226):
    C1=B9.find_element(By.ID,"formDetalhePortalTransparencia:frgDetalhePortalTransparencia")
    #C2=C1.find_element(By.ID,"formDetalhePortalTransparencia:frgPlanosResiduos")
    C3=C1.find_element(By.ID,"formDetalhePortalTransparencia:dt"+A)
    C4=C3.find_element(By.CLASS_NAME,"ui-datatable-tablewrapper")
    C5=C4.find_element(By.ID,"formDetalhePortalTransparencia:dt"+A+"_data")

    items2=C5.find_elements(By.CLASS_NAME,"ui-widget-content.ui-datatable-even.ui-datatable-selectable")
    items3=C5.find_elements(By.CLASS_NAME,"ui-widget-content.ui-datatable-odd.ui-datatable-selectable")

   
    for item in items2:
        texto=item.text.split()[0]
        if(ANO in texto):
            LEADS.append(item.text)
    
    for a in items3:
        texto=a.text.split()[0]
        if(ANO in texto):
            LEADS.append(item.text)
    
    D1=C3.find_element(By.ID,"formDetalhePortalTransparencia:dt"+A+"_paginator_top")
    D2=D1.find_element(By.CLASS_NAME,"ui-paginator-next.ui-state-default.ui-corner-all")
    Button4=D2.find_element(By.CLASS_NAME,"ui-icon.ui-icon-seek-next")
    
    C1=[]
    C2=[]
    C3=[]
    C4=[]
    C5=[]
    items2=[]
    items3=[]
    texto=[]

    Button4.click()
    time.sleep(5)    
```

```python
df = pd.DataFrame(data=LEADS, columns=["Descricao"])
df= df.drop_duplicates()
df.to_csv(LOCAL,sep=";",decimal=",",encoding="iso-8859-1",index=False)
```
